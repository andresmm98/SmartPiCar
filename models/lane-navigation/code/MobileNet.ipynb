# Prepare data

```python
train_dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir + '/manual-images-v2', 
    labels=steering_angles, 
    label_mode='int', 
    color_mode='rgb',
    batch_size=64,
    image_size=(240,320),
    seed=123)

test_dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir + '/validation', 
    labels=ysteering_angles, 
    label_mode='int', 
    color_mode='rgb',
    batch_size=64,
    image_size=(240,320),
    seed=123) 
```

# Base model

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(224,224,3),
                                               include_top=False,
                                               weights='imagenet')

print('Layers of the base model: ', len(base_model.layers))

base_model.trainable = False
```

# Data preprocessing

```python
preprocessing_layers = tf.keras.Sequential([
  tf.keras.layers.Cropping2D(cropping=((120,0),(0,0))),
  tf.keras.layers.Resizing(224,224),
  tf.keras.layers.Rescaling(1./127.5,offset=-1)])
```

# Regression head

```python
model = tf.keras.Sequential([
  tf.keras.layers.Cropping2D(cropping=((120,0),(0,0))),
  tf.keras.layers.Resizing(224,224),
  tf.keras.layers.Rescaling(1./127.5,offset=-1),
  base_model,
  tf.keras.layers.Conv2D(filters=32, kernel_size=3,activation='relu'),
  tf.keras.layers.GlobalAveragePooling2D(),
  tf.keras.layers.Dense(1)
])
```

# Configure model

```python
model.compile(loss='mse', 
              optimizer='adam')
```

```python
model.build((64,240,320,3))
model.summary()
```

# Training

```python
checkpoint_callback = keras.callbacks.ModelCheckpoint(filepath=os.path.join(model_output_dir,'autonomous_car_3_check.h5'), verbose=1, save_best_only=True)

history = model.fit(train_dataset,
                    epochs=10,
                    verbose=1,
                    callbacks=[checkpoint_callback],
                    validation_data=test_dataset)
```

## Save model

```python
model.save(os.path.join(model_output_dir,'lane_navigation_MobileNet.h5'))
```

# Print accuracy

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()
ax.plot(loss[:10], label='Training Loss')
ax.plot(val_loss[:10], label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Mean Squared Error')
plt.title('MobileNet V2')
plt.xlabel('epoch',labelpad=2)
ax.yaxis.tick_right()
plt.show()
```
